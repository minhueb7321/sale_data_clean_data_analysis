```python
def check_info_df(data):
    dataframe = pd.DataFrame(index = data.columns)
    dataframe['date_type'] = data.dtypes
    dataframe['count_unique'] = data.nunique()
    dataframe['not_null_value'] = data.count()
    dataframe['null_value'] = data.isnull().sum()
    dataframe['percentage_of_null'] = data.isnull().sum().mean()*100
    return dataframe
```
